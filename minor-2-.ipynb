## Emotion Recognition

```python
import cv2
import matplotlib.pyplot as plt
from deepface import DeepFace
```

```python
image_path ='Data/Humans/1 (1016).jpg'
image = cv2.imread(image_path)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
count = 0
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
faces = faceCascade.detectMultiScale(
        gray,
        scaleFactor=1.3,
        minNeighbors=3,
        minSize=(30, 30)
) 

if(faces.any()):
    count +=1

faceCascade_alt1 = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_alt.xml")
faces_alt1 = faceCascade.detectMultiScale(
        gray,
        scaleFactor=1.3,
        minNeighbors=3,
        minSize=(30, 30)
) 


if(faces_alt1.any()):
    count +=1
    
faceCascade_alt2 = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_alt2.xml")
faces_alt2 = faceCascade.detectMultiScale(
        gray,
        scaleFactor=1.3,
        minNeighbors=3,
        minSize=(30, 30)
) 


if(faces_alt2.any()):
    count +=1
```

```python
print("Found {0} Faces with default".format(len(faces)))
print("Found {0} Faces with alt 1".format(len(faces_alt1)))
print("Found {0} Faces with alt 2".format(len(faces_alt2)))
```

```python
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0),5)
```

```python
plt.imshow(image)
```

```python
if((len(faces) or len (faces_alt1) or (faces_alt2) >=1)):
    img1 = cv2.imread(image_path)
    results = DeepFace.analyze(img1 , actions =['emotion'])
    print(results)
    emotion = results["dominant_emotion"]
    print("\n\n\n\n The person is ",emotion)
    
```
